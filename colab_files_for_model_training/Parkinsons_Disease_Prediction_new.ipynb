
Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Analysis

```python

# loading the data from csv file to a Pandas DataFrame
parkinsons_data = pd.read_csv('/content/parkinsons.csv')
```

```python
# printing the first 5 rows of the dataframe
parkinsons_data.head()
```

```python

# number of rows and columns in the dataframe
parkinsons_data.shape
```

```python
# getting more information about the dataset
parkinsons_data.info()
```

```python

# checking for missing values in each column
parkinsons_data.isnull().sum()
```

```python

# getting some statistical measures about the data
parkinsons_data.describe()
```

```python


# distribution of target Variable
parkinsons_data['status'].value_counts()
```

```python
# grouping the data based on the target variable

# parkinsons_data.groupby('status').mean()

# or

# Group by 'status' and calculate the mean only for numeric columns

numeric_columns = parkinsons_data.select_dtypes(include='number').columns
parkinsons_data.groupby('status')[numeric_columns].mean()
```

Data Pre-Processing

Separating the features & Target

```python

X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python

print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
# training the SVM model with training data
model.fit(X_train, Y_train)

# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)

# accuracy score on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
y_pred = model.predict(X_test)  # Predicted labels
y_true = Y_test


# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='binary')  # Use 'macro' or 'micro' for multi-class
recall = recall_score(y_true, y_pred, average='binary')  # Use 'macro' or 'micro' for multi-class
f1 = f1_score(y_true, y_pred, average='binary')  # Use 'macro' or 'micro' for multi-class

# Print the metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Create a heatmap for visualization
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create and train the Decision Tree model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, Y_train)

# Accuracy score on training data
X_train_prediction = decision_tree_model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Decision Tree - Accuracy score of training data : ', training_data_accuracy)

# Accuracy score on testing data
X_test_prediction = decision_tree_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Decision Tree - Accuracy score of test data : ', test_data_accuracy)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create and train the KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors (k)
knn_model.fit(X_train, Y_train)

# Accuracy score on training data
X_train_prediction = knn_model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('KNN - Accuracy score of training data : ', training_data_accuracy)

# Accuracy score on testing data
X_test_prediction = knn_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('KNN - Accuracy score of test data : ', test_data_accuracy)
```
