 Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python

# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python

# printing the first 5 rows of the dataset
diabetes_dataset.head()
```

```python

# number of rows and Columns in this dataset
diabetes_dataset.shape
```

```python

# getting the statistical measures of the data
diabetes_dataset.describe()
```

```python

diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python

# separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python

print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

svm with k folds

```python

from sklearn import svm
from sklearn.metrics import accuracy_score

# Defining the model
svm_model = svm.SVC(kernel='linear')

# Setting up StratifiedKFold Cross Validation
k = 5  # Number of folds
skf = StratifiedKFold(n_splits=k)

# Performing cross-validation
cross_val_accuracies = cross_val_score(svm_model, X, Y, cv=skf, scoring='accuracy')

# Printing cross-validation results
print(f"Cross-Validation Accuracies for {k}-folds: {cross_val_accuracies}")
print(f"Mean Accuracy: {cross_val_accuracies.mean()}")

# Fit the SVM model on the entire dataset
svm_model.fit(X, Y)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Training the Logistic Regression model
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train, Y_train)

# Evaluating the model on the training data
X_train_prediction = log_reg_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Logistic Regression - Training Accuracy:', training_data_accuracy)

# Evaluating the model on the test data
X_test_prediction = log_reg_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Logistic Regression - Test Accuracy:', test_data_accuracy)
```

K-Nearest Neighbors

```python

from sklearn.neighbors import KNeighborsClassifier

# Training the KNN model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)

# Evaluating the model on the training data
X_train_prediction = knn_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('K-Nearest Neighbors - Training Accuracy:', training_data_accuracy)

# Evaluating the model on the test data
X_test_prediction = knn_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('K-Nearest Neighbors - Test Accuracy:', test_data_accuracy)
```

XGBoost Classifier

```python

from xgboost import XGBClassifier

# Training the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_train, Y_train)

# Evaluating the model on the training data
X_train_prediction = xgb_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('XGBoost - Training Accuracy:', training_data_accuracy)

# Evaluating the model on the test data
X_test_prediction = xgb_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('XGBoost - Test Accuracy:', test_data_accuracy)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Training the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=2)
rf_model.fit(X_train, Y_train)

# Evaluating the model on the training data
X_train_prediction = rf_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Random Forest - Training Accuracy:', training_data_accuracy)

# Evaluating the model on the test data
X_test_prediction = rf_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Random Forest - Test Accuracy:', test_data_accuracy)
```

Gradient Boost

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Training the Gradient Boosting model
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=2)
gb_model.fit(X_train, Y_train)

# Evaluating the model on the training data
X_train_prediction = gb_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Gradient Boosting - Training Accuracy:', training_data_accuracy)

# Evaluating the model on the test data
X_test_prediction = gb_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Gradient Boosting - Test Accuracy:', test_data_accuracy)
```
